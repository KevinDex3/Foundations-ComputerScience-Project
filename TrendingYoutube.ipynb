```python
# Environment
import pandas as pd
```

# 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
# Use of ecoding_erorrs to handle characters that cannot be decoded, ensuring the data is loaded correctly while marking invalid characters
CAvideos = pd.read_csv("CAvideos.csv")
DEvideos = pd.read_csv("DEvideos.csv")
FRvideos = pd.read_csv("FRvideos.csv")
GBvideos = pd.read_csv("GBvideos.csv")
INvideos = pd.read_csv("INvideos.csv")
JPvideos = pd.read_csv("JPvideos.csv",encoding_errors="replace")
KRvideos = pd.read_csv("KRvideos.csv",encoding_errors="replace")
MXvideos = pd.read_csv("MXvideos.csv",encoding_errors="replace")
RUvideos = pd.read_csv("RUvideos.csv",encoding_errors="replace")
USvideos = pd.read_csv("USvideos.csv")
```

```python
CAvideos
```

```python
# Add the column country

CAvideos["country"] = "CA"
DEvideos["country"] = "DE"
FRvideos["country"] = "FR"
GBvideos["country"] = "GB"
INvideos["country"] = "IN"
JPvideos["country"] = "JP"
KRvideos["country"] = "KR"
MXvideos["country"] = "MX"
RUvideos["country"] = "RU"
USvideos["country"] = "US"

# Concatenate all the data frames
df_all = pd.concat([CAvideos, DEvideos, FRvideos, GBvideos, INvideos,JPvideos, KRvideos, MXvideos, RUvideos, USvideos], ignore_index=True)

df_all
```

# 2. Extract all videos that have no tag

```python
# Videos that have no tag

no_tags = df_all[df_all["tags"] == "[none]"]
no_tags
```

# 3. For each channel, determine the total number of views

```python
df_all.groupby("channel_title")["views"].sum().reset_index()
```

# 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
excluded = df_all[
    (df_all["comments_disabled"] == True) |
    (df_all["ratings_disabled"] == True) |
    (df_all["video_error_or_removed"] == True)
]

excluded
```

```python
df_all = df_all[
    (df_all["comments_disabled"] == False) |
    (df_all["ratings_disabled"] == False) |
    (df_all["video_error_or_removed"] == False)
]

df_all
```

# 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
# Numbers of video with dislikes = 0

(df_all["dislikes"] == 0).sum()
```

```python
# Use of replace because we have 9763 values with dislikes = 0 (division by zero)

df_all["like_ratio"] = df_all["likes"]/df_all["dislikes"].replace(0, 1)

df_all
```

# 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
# Convert the 'publish_time' column from string to datetime
df_all['publish_time'] = pd.to_datetime(df_all['publish_time'])
df_all
```

```python
# Define start which contains the time rounded to the beginning of the interval, and end which contains the time rounded to the end of the interval

start = df_all["publish_time"].dt.floor("10min")
end = start + pd.Timedelta(minutes=10)

# Creation of new variable "interval_10min"

df_all["interval_10min"] = (start.dt.strftime("%H:%M") + " - " + end.dt.strftime("%H:%M"))

df_all
```

# 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
# For each interval the number of videos

num_of_videos = df_all.groupby("interval_10min")["video_id"].count()
num_of_videos
```

```python
print("The max number is: " + str(max(num_of_videos)) + " and the min number is " + str(min(num_of_videos)))
```

```python
# For each interval the average number of likes

average_number_likes = df_all.groupby("interval_10min")["likes"].mean().reset_index()
average_number_likes
```

```python
# For each interval the average number of dislikes
average_number_dilikes = df_all.groupby("interval_10min")["dislikes"].mean().reset_index()
average_number_dilikes
```

# 8. For each tag, determine the number of videos

```python
# Transform into a list

df_all["tags_list"] = df_all["tags"].str.split("|")
df_all
```

```python
df_exploded = df_all.explode("tags_list")
df_exploded
```

```python
# For each tag, the number of videos

tag_counts = df_exploded.groupby("tags_list")["video_id"].count().reset_index()
tag_counts.columns = ["tag", "num_videos"]                                          # Rename video_id in num_videos
tag_counts
```
